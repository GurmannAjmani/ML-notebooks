# ZERO SHOT CLASSIFICATION= CLASSIFIES ACCORDING TO CATEGORIES

```python
zero=pipeline("zero-shot-classification")
```

```python
zero("I am a student of politics",candidate_labels=["education","politics"])
```

```python
result=zero("I am a businessman",candidate_labels=["education","politics","business"])
```

```python
print(result["scores"])
```

# Question answering with context

```python
answer=pipeline("question-answering")
```

```python
answer(question="What is my friends name?",context="My name is Gurmann and my friends name is John Doe")
```

# BERT 

```python
unmasker = pipeline("fill-mask", model="bert-base-uncased")
```

```python
result = unmasker("This man works as a [MASK].")
print(result)
```

```python
result = unmasker("I work at the[MASK].")
print(result)
```
